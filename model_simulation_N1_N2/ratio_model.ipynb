```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
from model_2x2_const_Y import get_time_series_N1_N2_ratio_Y
from model_2x2_const_X import get_time_series_N1_N2_ratio_X
```

```python
lst_timex, lst_N1x, lst_N2x, lst_ratiox = get_time_series_N1_N2_ratio_X(N1=1000, N2=500, total_time=2000)
lst_timey, lst_N1y, lst_N2y, lst_ratioy = get_time_series_N1_N2_ratio_Y(N1=1000, N2=500, total_time=2000)
```

```python
plt.plot(lst_timex, lst_N1x, label="N1")
plt.plot(lst_timex, lst_N2x, label="N2")
plt.xlabel("time")
plt.ylabel("N1 N2 X")
plt.legend()
plt.savefig("N1_N2_model_2x2_const_Y.png")
```

```python
plt.plot(lst_timey, lst_N1y, label="N1")
plt.plot(lst_timey, lst_N2y, label="N2")
plt.xlabel("time")
plt.ylabel("N1 N2 Y")
plt.legend()
plt.savefig("N1_N2_model_2x2_const_Y.png")
```

```python
ratio_dfx = pd.Series(lst_ratiox, index=lst_timex)
plt.plot(lst_timex, lst_ratiox, label="ratio_N1_N2")
plt.xlabel("time")
plt.ylabel("ratio")
plt.legend()
plt.savefig("ratio_model_2x2_const_X.png")
plt.show()
```

```python
ratio_dfy = pd.Series(lst_ratioy, index=lst_timey)
plt.plot(lst_timey, lst_ratioy, label="ratio_N1_N2")
plt.xlabel("time")
plt.ylabel("ratio")
plt.legend()
plt.savefig("ratio_model_2x2_const_Y.png")
plt.show()
```

```python
plt.figure(figsize=(16, 8))
ratio_dfx.plot(style="k--")
ratio_df_moving_avg = ratio_dfx.rolling(5000).mean()
ratio_df_moving_std = ratio_dfx.rolling(5000).std()
ratio_df_moving_avg.plot(label="ratio moving avg", color="red")
ratio_df_moving_std.plot(style="b", label="ratio moving std")
plt.plot(ratio_dfx.index, [1.5]*len(ratio_dfx.index), label="y=1.5")
plt.savefig("ratio_model_2x2_const_Y_w_mean_std.png")
plt.xlabel("Time")
plt.ylabel("N1/N2 ratio")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 8))
ratio_dfy.plot(style="k--")
ratio_df_moving_avg = ratio_dfy.rolling(5000).mean()
ratio_df_moving_std = ratio_dfy.rolling(5000).std()
ratio_df_moving_avg.plot(label="ratio moving avg", color="red")
ratio_df_moving_std.plot(style="b", label="ratio moving std")
plt.plot(ratio_dfy.index, [1.5]*len(ratio_dfy.index), label="y=1.5")
plt.savefig("ratio_model_2x2_const_Y_w_mean_std.png")
plt.xlabel("Time")
plt.ylabel("N1/N2 ratio")
plt.legend()
plt.show()
```
